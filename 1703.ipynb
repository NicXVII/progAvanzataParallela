# mmap- mappatura file memoria, comodo accedere contenuto file come fosse direttamente memoria- spazio indirizzi maggiore memoria fisica

```
/** * mmap - Mappa una regione di memoria a un file * * @param file - Indirizzo della memoria mappata (output) * @param offset - Offset all'interno dello spazio di indirizzamento del processo * @param length - Lunghezza in byte della regione da mappare * @param prot - Protezione della memoria (PROT_READ, PROT_WRITE, PROT_EXEC, PROT_NONE) * @param flags - Opzioni di mapping (MAP_SHARED, MAP_PRIVATE, MAP_FIXED, ecc.) * @param fd - File descriptor del file da mappare * @param file_offset - Offset all'interno del file da cui iniziare la mappatura * * @return - Indirizzo della memoria mappata al successo, MAP_FAILED (-1) in caso di errore * * Descrizione: * La funzione mmap() stabilisce una mappatura tra uno spazio di indirizzi di processo * e un file, permettendo di accedere al contenuto del file come se fosse memoria. * Questa tecnica e utile per operazioni di I/O ad alte prestazioni. */mmap(file, offset, length, prot, flags, fd, file_offset)
```

## come mappare- apro file, ottengo file descriptor, posso ottenere dimensione file## madvise- usato per indicare indirizzi da accedere## nmap### vantaggi- accesso diretto file, non seere reafm write, lseek, affido al sistema gestione memoria. piu facile per thred o processi accedere risorse### svantaggi- non sempre piu efficiente, mappare file macchinoso, nmap esiste solo POSIX, non posso accedere file dim sup spazio indirizzi, affido caching al sistema gestione memoria (non sempre efficiente)
