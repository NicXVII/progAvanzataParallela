generatore - interrompere e ripdenre esecuzione
    Costruiamo un generatore x = f()
    •  ottenere  primo valore con yield chiamiamo  funzione next  argomento  generatore, e.g., next(x)
    • La prossima volta che chiamiamo next l’esecuzione di f riprenderà dello yield in cui si era fermata, per
    proseguire fino allo yield successivo (o, se non c’è, produrre una eccezione di tipo StopIteration)
        

```python
def f():
    k = 1
    while True:
        yield k 
        k += 1

x = f()
print(next(x))
print(next(x))
```

iteratori: oggetti comportano come generatori
due metodi: __iter__(self) svolge inizializzazione e ritorna se stesso
            __next__(self) ritorna  prossimo elemento, o solleva eccezione StopIteration se non ci sono più elementi

```python
class Cubes:
    def __init__(self, num):
        self.num = num
        
    def __iter__(self):
        self.k = 0
        return self
    
    def __next__(self):
        if self.k > self.num:
            raise StopIteration
        else:
            res = self.k
            self.k += 1
            return res**3
        
    
cub = Cubes(10)
for x in cub:
    print(x)
```

```python
from functools import wraps
caches ={}

def cached(f):
    name = f.__name__
    if name not in caches:
        caches [name] = {}
    @wraps(f)
    def with_cache(x):
        if x in caches[name]:
            print("Cache hit")
            return caches[name][x]
        res = f(x)
        caches[name][x] = res
        return res
    return with_cache

@cached
def f(x):
    for i in range(100_000_000):
        x += 1
    return x

@cached
def g(x):
    for i in range(1,1000):
        x = x*1
    return x



f(5)


```
