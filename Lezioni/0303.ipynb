## esecuzione speculativa
- un branch predictor tiene statistiche sui brach presi/ no presi e usa per predizione
- bp statico non tiene statistiche, sceglie sempre preso o no preso

### branch predictor due livelli
- si usa lo storico n bit per indicizzare taella il predittore appoisto per quello storico
- locale. predittore separato per diversi branch
- globale. predittore unico per tutti i brancch utile per la correlazionee tra branch

## sequenze di compilazione
- preprocessing: espansione macro, inserimento header, rimozione dei commenti
    - gcc -E source.c /stampa standard output
- compilazione: parsing del sorgente, controllo errori sintassi, generazione codice assembly
    - gcc -S source.c /genera file assembly
- Assembly: trasforma assembly in codice macchina, tutte le chiamate a funzioni esterne sono psostituite da placeolder
    - gcc -c source.s /genera file source
- linking: risolve i placeholder con indirizzi reali, genera eseguibile
