# mmap

- mappatura file memoria, comodo accedere contenuto file come fosse direttamente memoria
- spazio indirizzi maggiore memoria fisica

```c
/**
 * mmap - Mappa una regione di memoria a un file
 *
 * @param file - Indirizzo della memoria mappata (output)
 * @param offset - Offset all'interno dello spazio di indirizzamento del processo
 * @param length - Lunghezza in byte della regione da mappare
 * @param prot - Protezione della memoria (PROT_READ, PROT_WRITE, PROT_EXEC, PROT_NONE)
 * @param flags - Opzioni di mapping (MAP_SHARED, MAP_PRIVATE, MAP_FIXED, ecc.)
 * @param fd - File descriptor del file da mappare
 * @param file_offset - Offset all'interno del file da cui iniziare la mappatura
 *
 * @return - Indirizzo della memoria mappata al successo, MAP_FAILED (-1) in caso di errore
 *
 * Descrizione:
 * La funzione mmap() stabilisce una mappatura tra uno spazio di indirizzi di processo
 * e un file, permettendo di accedere al contenuto del file come se fosse memoria.
 * Questa tecnica e utile per operazioni di I/O ad alte prestazioni.

 */
mmap(file, offset, length, prot, flags, fd, file_offset)
```

## come mappare

- apro file, ottengo file descriptor, posso ottenere dimensione file

## madvise

- usato per indicare indirizzi da accedere

## nmap

### vantaggi

- accesso diretto file, non seere reafm write, lseek, affido al sistema gestione memoria. piu facile per thred o processi accedere risorse

### svantaggi

- non sempre piu efficiente, mappare file macchinoso, nmap esiste solo POSIX, non posso accedere file dim sup spazio indirizzi, affido caching al sistema gestione memoria (non sempre efficiente)

## esempio 1

```c
# include < stdio .h >
# include < stdlib .h >
# include < sys / types .h >
# include < sys / mman .h >
# include < sys / stat .h >

int main ( int argc , char * argv [])
{
char * data ;
struct stat sbuf ;
FILE * fd = fopen (" mmap_demo ", "r");
int fdnum = fileno ( fd );

stat (" mmap_demo ", & sbuf );

data = mmap (( void *) 0, sbuf . st_size ,
PROT_READ , MAP_SHARED , fdnum , 0) ;

for ( int i = 0; i < 10; i ++) {
printf ("%c %c\n", data [i ], getc ( fd ));
}

munmap ( data , sbuf . st_size );

fclose ( fd );

return 0;
}
```

## esempio 2

```c
# include < stdio .h >
# include < stdlib .h >
# include < unistd .h >
# include < sys / types .h >
# include < sys / mman .h >
# include < sys / stat .h >

void read_vector ( int len )
{
FILE * fd = fopen (" mmap_vector_demo ", "r");
if ( fd == NULL ) {
printf (" Impossibile aprire il file \n");
return ;
}
int fdnum = fileno ( fd );
int * data = mmap (( void *) 0, len * sizeof (
int ) , PROT_READ , MAP_SHARED , fdnum , 0) ;
if ( data == MAP_FAILED ) {
printf (" Impossibile effettuare il mapping \n"
);
return ;
}
int sum = 0;
for ( int i = 0; i < len ; i ++) {
sum += data [i ];
}
printf ("La somma dei valori salvati nel file
e' %d\n", sum ) ;
munmap ( data , len * sizeof ( int ));
fclose ( fd );
}

void write_vector ( int len )
{
FILE * fd = fopen (" mmap_vector_demo ", "w+") ;
if ( fd == NULL ) {
printf (" Impossibile aprire il file \n");
return ;
}
int fdnum = fileno ( fd );
ftruncate ( fdnum , len * sizeof ( int ));

int * data = mmap (( void *) 0, len * sizeof (
int ) , PROT_READ | PROT_WRITE , MAP_SHARED ,
fdnum , 0) ;
if ( data == MAP_FAILED ) {
printf (" Impossibile effettuare il mapping \n"
);
return ;
}
for ( int i = 0; i < len ; i ++) {
data [i] = i;
}
munmap ( data , len * sizeof ( int ));
fclose ( fd );

}

int main ( int argc , char * argv [])
{
const int len = 10000;
write_vector ( len ) ;
read_vector ( len );
return 0;
}
```

Tassonomia Flynn 
- SISD: Single Instruction Single Data, un processore esegue una singola istruzione su un singolo dato alla volta.
- SIMD: Single Instruction Multiple Data, un processore esegue una singola istruzione
su più dati contemporaneamente, ad esempio, sommando elementi di due array in parallelo.
- MISD: Multiple Instruction Single Data, un processore esegue più istruzioni su un singolo dato, ma questa architettura è rara e non comune.
- MIMD: Multiple Instruction Multiple Data, un processore esegue più istruzioni su più dati contemporaneamente, ad esempio, in un sistema multi-core dove ogni core esegue istruzioni diverse su dati diversi
